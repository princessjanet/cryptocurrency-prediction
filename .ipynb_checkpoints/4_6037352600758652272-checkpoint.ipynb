```python
def dateconv(date):
  benchmark=0
  time=int(date.split("T")[-1].split(":")[0])
  DD=(int(date.split("T")[0].split("-")[1])*30 + int(date.split("T")[0].split("-")[2]))*24 +time
  return (DD)
```
